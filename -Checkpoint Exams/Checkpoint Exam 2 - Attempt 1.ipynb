**Question 1** \
Which of the following statements is *most accurate* about a 3-for1 stock split in one of the securities in a price-weighted index?

**Answer:** \
<b><font color="green">A) The relative weights of the other securities in the index will increase.
(Correct Answer)</font></b>
\
\
B) The index value will decrease because the split security’s price will decrease.
\
\
C) The index divisor will have to be increased to offset the price decrease of the split security.
\
\
**Explanation:** \
A split in one security will reduce its price and therefore its weight in the index. As a result, all other securities whose prices have not changed will increase their relative weight in the index. The index value will need to be *maintained* at the pre-split level. To maintain the index value, the index divisor will have to be adjusted down because the numerator is also adjusted down.
\
\
(Module 42.1, LOS 42.e)

**Question 7** \
Grovers Inc. manufactures children's toys in a country where it has a significant share of the market and barriers to entry are high. Proxe Corp. manufactures underwear globally in a highly competitive environment. Recent innovations in machinery have reduced operating expenses in both Govers's and Proxe's industries. The reduced operating expenses are *most likely* to:



**Answer:** \
A) increase operating margins for both companies.
\
\
<b><font color="green">B) increase operating margins more for Grovers than for Proxe. (Correct Answer)</font></b>
\
\
C) have little impact on the operating margins of either company.
\
\
**Explanation:** \
Because Grovers operates in a market with weaker competitive forces than Proxe, Grovers likely has more pricing power than Proxe. **Therefore**, Grovers can likely keep its pricing unchanged so that its lower operating costs will increase its operating margins. By contrast, in Proxe's market, competitors are more likely to respond to decreased operating expenses by decreasing their prices and attempting to gain market share. Proxe will likely need to decrease its prices to defend its market share, which will keep pressure on its operating margins.
\
\
(Module 46.2, LOS 46.d) and (Module 47.1, LOS 47.c)

**Question 8** \
AgroCorp Inc., a Canadian grain producer, recently issued a five-year floating rate note with a face value of CAD 90,000,000. The bond pays semiannual interest based on the Canadian Overnight Repo Rate (CORRA) plus 150 bps per year. If CORRA is 0.50% in the final interest period, at maturity AgroCorp will pay bondholders:


**Answer:** \
A) CAD 1,800,000.
\
\
<b><font color="green">B) CAD 90,900,000. (Correct Answer)</font></b>
\
\
C) CAD 91,800,000.
\
\
**Explanation:** \
At maturity, AgroCorp will have to pay both the final coupon and the principal value of the bond to investors. The coupon rate is `CORRA + 0.50% = 2.00%` annually. For the last six-month period, 
\
the coupon interest is `2% / 2 × CAD 90,000,000 = CAD 900,000`. 
\
Therefore, AgroCorp will have to pay bondholders `CAD 90,000,000 + CAD900,000 = CAD 90,900,000.`
\
\
(Module 46.2, LOS 46.d) and (Module 47.1, LOS 47.c)

**Question 11** \

**Answer:** \
A) par.
\
\
B) a discount.
\
\
<b><font color="green">C) a premium. (Correct Answer)</font></b>

\

\

**Question 12** \
A fixed income trader expects that the yield of a 5-year, 5% annual coupon bond with a Macaulay duration of 4.25 will <u>decrease from 7% to 6%</u>. As a result, the anticipated percentage increase in the bond's price is:


**Answer:** \
A) 4.25%.
\
\
<b><font color="green">B) less than 4.25%. (Correct Answer)</font></b>
\
\
C) more than 4.25%..


**Explanation:** \
We can estimate the bond's price change based on modified duration. We can calculate modified duration by dividing Macaulay duration by the yield:
$$
\text{ModDur } = \large \frac{\text{MacDur}}{1 + r} \small =  \large \frac{4.25}{1.06} = 4.009
$$
\
\
As a result, the estimated price increase of the bond for a `100 bps decrease in yield is 4.009%`.


(Module 59.1, LOS 59.a)

**Question 18** \
A sequential-pay collateralized mortgage obligation (CMO) has three tranches: Tranche A matures in 2 years, Tranche B matures in 5 years, and Tranche C matures in 10 years. An investor who expects a gradual decline in interest rates over the next 5 years would *most likely* be concerned about:



**Answer:** \
A) contraction risk, and therefore would prefer Tranche A.
\
\
B) extension risk, and therefore would prefer Tranche B.
\
\
<b><font color="green">C) contraction risk, and therefore would prefer Tranche C. (Correct Answer)</font></b>

\

The investor would be concerned about contraction risk, or the risk that decreasing interest rates will cause interest payments to decline and principal to be prepaid more quickly than anticipated over the next 5 years. \

Which of the following reasons best explains why BamBam should use a derivative rather than a spot market transaction to manage the currency risk of the contract?
