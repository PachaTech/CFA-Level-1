### Quantitative Methods

$FV = PV (1 + r)^t$
\
\
$PV = \Large\frac{FV}{(1 + r)}^t = FV(1 + r)^{–t}$
\
\
where:

$r$ = interest rate per compounding period

$t$ = number of compounding periods

<p></p>

**continuous compounding:**

$FV = PV × e^{rt}$

$PV = FV × e^{–rt}$

PV of a perpetuity = $\Large\frac{\text{payment}}{r}$
\
\
annuity payment = $\Large\frac{r × PV}{1 – (1 + r)^{–t}}$

where:

$r$ = interest rate per period

$t$ = number of periods

$PV$ = present value (principal)

<p></p>


constant growth dividend discount model: $V_0 = \Large\frac{D_1}{k_e−g_c}$
\
\
required rate of return: $\Large k_e = \frac{D_1}{V_0} \small + \large g_c$
\
\
implied growth rate: $\Large gc = k_e - \frac{D_1}{V_0}$
\
\
sample mean:  
\
$\qquad \Large \bar{X} = \Large\frac{\sum_{i=1}^{n} X_i}{n}$
\
\
range = maximum value − minimum value
\
\
mean absolute deviation = $\Large\frac{\sum_{i=1}^{n} ∣X_i−X∣}{n}$
\
\
sample variance: 
\
\
\
$s^2 = \Large\frac{\sum_{i=1}^{n}(X_i−\overline{X})^2}{n−1}$
\
\
\
sample standard deviation:
\
\
$s = \Large \sqrt{\frac{\sum_{i=1}^{n}(X_i−\overline{X})^2}{n−1}}$
\
\
\
\
coefficient of variation=
\
\
$\Large\frac{S_x}{\overline{X}} = \frac{\text{standard deviation of x}}{\text{average value of x}}$
\
\
\
target downside deviation: 
\
\
$s_\text{target} = \Large\sqrt{\frac{\sum_{\text{all} X_i < B}^{n}(X_i–B)^2} {n−1}}$
\
\
where $B$ = target value
\
\
\
\
sample covariance: 
\
\
$\quad S_{XY} = \Large \frac{\sum_{i=1}^n [X_i−\overline{X}][Yi−\overline{Y}]}{n−1}$
\
\
\
where:

$X_i$ = an observation of variable X

$Y_i$ = an observation of variable Y

$\overline{X}$ = mean of variable X

$\overline{Y}$ = mean of variable Y

$n$ = number of periods

<p></p>

where:

$\overline{Y}$= mean of Y

<p></p>

\
\
total sum of squares (SST): 
\
\
$\Large SST=\sum_{i=1}^n (Y_i − \overline{Y})^2$

<p></p>

sum of squares regression (SSR):  
\
$\Large SSR=\sum_{i=1}^n (\hat{Y}_i − \overline{Y})^2$

<p></p>

\
\
$\Large SSE=\sum_{i=1}^n (Y_i-\hat{Y})^2$


<p></p>

\
\
$R^2 = \Large\frac{SSR}{SST}$

rate of return for an investor after fees:

$r = \Large\frac{V_1 - V_0 - \text{total fees}}{V0} $

<p></p>

$\qquad r ×(V_0 + V_B) - (r_B × V_B)$

$\qquad \text{total fees} = m V_1 + \max[0, p(V_1 - V_0)]$
